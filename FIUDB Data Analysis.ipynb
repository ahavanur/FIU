```python
ctr_df['dateOfTransaction'] = pd.to_datetime(ctr_df['dateOfTransaction'])
def check(x):
    try:
        return x.days
    except:
        return 0
ctr_df['DaysSince'] = pd.Series(map(lambda x: check(x), ctr_df['dateOfTransaction'] - min(ctr_df['dateOfTransaction'])))
ctr_df['cashAmount'] = pd.Series(list(map(lambda x: float(x.replace("$",'')) if type(x) == str else x, ctr_df['cashAmount'])))
plt.hist(ctr_df['cashAmount'])
plt.show()
```
